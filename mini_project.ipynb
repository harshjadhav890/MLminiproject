```python
import matplotlib.pyplot as plt
import pandas as pd
from skimage.io import imread
from skimage.transform import resize 
from skimage.color import rgb2gray
```

```python
# dataset path - "/content/drive/MyDrive/2. My Stuff/SmartKnower data science/Mini Project/dataset"
```

```python
import os 
import numpy as np

datadir = '/content/drive/MyDrive/2. My Stuff/SmartKnower data science/Mini Project/dataset'
types = ['football', 'monkey']
flat_array = []
target = []

for i in types:
  path = os.path.join(datadir,i)
  for j in os.listdir(path):
    img = os.path.join(path,j)
    img_array = imread(img)
    # Preprocessing
    # Converting to Grayscale
    img_resize = rgb2gray(resize(img_array,(150,150)))
    flat_array.append(np.ndarray.flatten(img_resize))
    target.append(i)

flat_array = np.array(flat_array)
target = np.array(target)
```

```python
df = pd.DataFrame(flat_array)
df['Target'] = target
df.head()
```

```python
x = flat_array
y = target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=0,stratify=y) 
```

```python
np.unique(y_test, return_counts=True)
```

```python
# hyperparameter tuning

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
model = SVC()
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(x_train, y_train)
```

```python
print(grid.best_params_)
```

```python
# Creation of model

model = SVC(C=1,gamma=0.0001,kernel='rbf')
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
# Printing Metrics

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_score(y_pred,y_test)
print(classification_report(y_pred,y_test))
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```
